**Phần 5: Mô tả dữ liệu và khảo sát dạng phân phối**

1. Hãy mô tả và khảo sát phân phối cho biến T1

```python
t1_stats = df['T1'].describe()
print(t1_stats)

# Vẽ biểu đồ Box-Plot và xác định các 10 đại lượng trong biểu đồ đó
import matplotlib.pyplot as plt
df['T1'].plot.box()
plt.show()

# Xác định các 10 đại lượng trong biểu đồ Box-Plot
# 1. Giá trị nhỏ nhất (Q1 - 1.5 * IQR)
q1 = t1_stats['25%']
iqr = t1_stats['75%'] - t1_stats['25%']
min_value = q1 - 1.5 * iqr

# 2. Giá trị phần tử đầu tiên (Q1 - 1.5 * IQR)
first_value = df['T1'].min()

# 3. Giá trị phần tử thứ 25% (Q1)
second_value = q1

# 4. Giá trị trung bình (Mean)
mean_value = t1_stats['mean']

# 5. Giá trị phần tử thứ 50% (Median)
median_value = t1_stats['50%']

# 6. Giá trị phần tử thứ 75% (Q3)
third_value = t1_stats['75%']

# 7. Giá trị phần tử thứ 95% (Q3 + 1.5 * IQR)
fourth_value = q3 + 1.5 * iqr

# 8. Giá trị phần tử cuối cùng (Q3 + 1.5 * IQR)
last_value = df['T1'].max()

# 9. Giá trị phần tử thứ 97.5% (Q3 + 3 * IQR)
fifth_value = q3 + 3 * iqr

# 10. Giá trị lớn nhất (Q3 + 3 * IQR)
max_value = q3 + 3 * iqr

# In ra các giá trị đã xác định
print("Giá trị nhỏ nhất:", min_value)
print("Giá trị phần tử đầu tiên:", first_value)
print("Giá trị phần tử thứ 25%:", second_value)
print("Giá trị trung bình:", mean_value)
print("Giá trị phần tử thứ 50%:", median_value)
print("Giá trị phần tử thứ 75%:", third_value)
print("Giá trị phần tử thứ 95%:", fourth_value)
print("Giá trị phần tử cuối cùng:", last_value)
print("Giá trị phần tử thứ 97.5%:", fifth_value)
print("Giá trị lớn nhất:", max_value)
```

```python
# Tính toán độ lệch chuẩn
std_dev = t1_stats['std']

# Tính toán độ lệch
skewness = (3 * (mean_value - median_value)) / std_dev

# Tính toán độ nhọn
kurtosis = (4 * (mean_value - median_value)) / std_dev

# Mô tả hình dáng lệch của phân phối T1
if skewness > 0:
  print("Phân phối lệch phải")
elif skewness < 0:
  print("Phân phối lệch trái")
else:
  print("Phân phối đối xứng")

# Vẽ biểu đồ Histogram biểu thị hình dáng phân phối
df['T1'].plot.hist(bins=10)
plt.show()
```

```python
# Mô tả các đặc trưng của phân phối
print("Phân phối có đặc điểm:")
print("- Giá trị trung bình:", df['T1'].mean())
print("- Độ lệch chuẩn:", df['T1'].std())
print("- Độ lệch:", skewness)
print("- Độ nhọn:", kurtosis)

# Mô tả mức độ lệch
if skewness > 0:
  print("Phân phối lệch phải, có nghĩa là các giá trị cao hơn có nhiều khả năng xảy ra hơn các giá trị thấp hơn.")
elif skewness < 0:
  print("Phân phối lệch trái, có nghĩa là các giá trị thấp hơn có nhiều khả năng xảy ra hơn các giá trị cao hơn.")
else:
  print("Phân phối đối xứng, có nghĩa là các giá trị cao hơn và thấp hơn có khả năng xảy ra như nhau.")

# Mô tả mức độ nhọn
if kurtosis > 0:
  print("Phân phối nhọn hơn phân phối chuẩn, có nghĩa là các giá trị gần giá trị trung bình có nhiều khả năng xảy ra hơn các giá trị xa giá trị trung bình.")
elif kurtosis < 0:
  print("Phân phối tù hơn phân phối chuẩn, có nghĩa là các giá trị xa giá trị trung bình có nhiều khả năng xảy ra hơn các giá trị gần giá trị trung bình.")
else:
  print("Phân phối có độ nhọn tương tự như phân phối chuẩn.")
```

```python
import scipy.stats as stats

# Kiểm chứng phân phối chuẩn QQ-Plot
stats.probplot(df['T1'], dist="norm", plot=plt)
plt.show()

# Nhận xét và đánh giá về phân phối của T1
# Dựa trên biểu đồ QQ-Plot, phân phối của T1 gần với phân phối chuẩn.
# Tuy nhiên, có một số điểm dữ liệu nằm ngoài đường chéo, điều này cho thấy có một số giá trị ngoại lai trong dữ liệu.

# Vẽ biểu đồ Histogram biểu thị hình dáng phân phối
df['T1'].plot.hist(bins=10)
plt.show()

# Mô tả các đặc trưng của phân phối
print("Phân phối có đặc điểm:")
print("- Giá trị trung bình:", df['T1'].mean())
print("- Độ lệch chuẩn:", df['T1'].std())
print("- Độ lệch:", skewness)
print("- Độ nhọn:", kurtosis)

# Mô tả mức độ lệch
if skewness > 0:
  print("Phân phối lệch phải, có nghĩa là các giá trị cao hơn có nhiều khả năng xảy ra hơn các giá trị thấp hơn.")
elif skewness < 0:
  print("Phân phối lệch trái, có nghĩa là các giá trị thấp hơn có nhiều khả năng xảy ra hơn các giá trị cao hơn.")
else:
  print("Phân phối đối xứng, có nghĩa là các giá trị cao hơn và thấp hơn có khả năng xảy ra như nhau.")

# Mô tả mức độ nhọn
if kurtosis > 0:
  print("Phân phối nhọn hơn phân phối chuẩn, có nghĩa là các giá trị gần giá trị trung bình có nhiều khả năng xảy ra hơn các giá trị xa giá trị trung bình.")
elif kurtosis < 0:
  print("Phân phối tù hơn phân phối chuẩn, có nghĩa là các giá trị xa giá trị trung bình có nhiều khả năng xảy ra hơn các giá trị gần giá trị trung bình.")
else:
  print("Phân phối có độ nhọn tương tự như phân phối chuẩn.")
```

```python
#Hãy khảo sát tương quan giữa biến DH1 theo biến T1 trên từng nhóm khu vực
for kv in df['KV'].unique():
  df_kv = df[df['KV'] == kv]
  correlation = df_kv['DH1'].corr(df_kv['T1'])
  print(f"The correlation between DH1 and T1 for KV {kv} is {correlation:.2f}")

  # Create the scatter plot
  plt.scatter(df_kv['T1'], df_kv['DH1'])

  # Add labels and title
  plt.xlabel('T1')
  plt.ylabel('DH1')
  plt.title(f'Scatter Plot of DH1 vs T1 for KV {kv}')

  # Show the plot
  plt.show()
```
