```python
import pandas as pd
import numpy as np
data = pd.read_csv("dulieuxettuyendaihoc.csv", index_col="STT")
df = pd.DataFrame(data = data)
print(df.head(10))
df.tail(10)
```

```python
df['DT'].isnull().value_counts()
print(df['DT'].isnull().value_counts())
```

```python
df['DT'] = df['DT'].fillna(0)
print(df['DT'].isnull().value_counts())
```

```python
# 4
df['DT'].fillna(0, inplace= False) # False không thay đổi dữ liệu gốc
```

```python
df['T1'].isnull().value_counts()
print(df['T1'].isnull().value_counts())
```

```python
# 1.  Lập bảng tần số, tần suất để khảo sát dữ liệu thiếu
missing_counts = df['T1'].isnull().value_counts()
print(missing_counts)

missing_table = pd.crosstab(df['T1'].isnull(), columns = 'count')
print(missing_table)

# 2. Thực hiện thay thế dữ liệu thiếu bằng phương pháp Mean
mean_t1 = df['T1'].mean()
df['T1'] = df['T1'].fillna(mean_t1)

# Kiểm tra lại dữ liệu còn thiếu
missing_counts = df['T1'].isnull().value_counts()
print(missing_counts)
```

```python
# 5
df['T1'].fillna(df['T1'].mean(), inplace= False)
```

```python
# 6
is_float64 = df.dtypes == float
df_column_list = df.columns[is_float64].tolist()
print(df_column_list)

for column in df_column_list:
    df[column].fillna(df[column].mean(), inplace= False)

df.head()
```

```python
# 7 Công thức tính: TBM = (T1*2 + L1 + H1 + S1 + V1*2 + X1 + D1 + N1) / 10

df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10
df.head()
```

```python
# 8
def Xeploai(TBM):
  if TBM < 5.0:
    return 'Y'
  elif 5.0 <= TBM < 6.5:
    return 'TB'
  elif 6.5 <= TBM < 8.0:
    return 'K'
  elif 8.0 <= TBM < 9.0:
    return 'G'
  else:
    return 'XS'

df['XL1'] = df['TBM1'].apply(Xeploai)
df['XL2'] = df['TBM2'].apply(Xeploai)
df['XL3'] = df['TBM3'].apply(Xeploai)

df.head()
```

```python
# 9
def min_max_normalization(df, column):
  minimum = df[column].min()
  maximum = df[column].max()
  normalized_values = (df[column] - minimum) / (maximum - minimum) * 4
  return normalized_values

df['US_TBM1'] = min_max_normalization(df, 'TBM1')
df['US_TBM2'] = min_max_normalization(df, 'TBM2')
df['US_TBM3'] = min_max_normalization(df, 'TBM3')

df.head()
```

```python
# 10
def KQXT(df):
  if df['KT'] in ['A', 'A1']:
    return 1 if (df['DH1'] * 2 + df['DH2'] + df['DH3']) / 4 >= 5.0 else 0
  elif df['KT'] == 'B':
    return 1 if (df['DH1'] * 2 + df['DH2'] + df['DH3']) / 4 >= 5.0 else 0
  else:
    return 1 if (df['DH1'] * 2 + df['DH2'] + df['DH3']) / 3 >= 5.0 else 0

df['KQXT'] = df.apply(KQXT, axis=1)
df.head()
```

```python
# Save to csv
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
```
