```python
column_names = ["ID", "Name", "Age" ,"Weight", "m0006", "m0612", "m1218", "f0006", "f0612", "f1218"]
patientheartrate = pd.read_csv(r'/content/patient_heart_rate.csv', names=column_names)
print(patientheartrate.head())
```

```python
patientheartrate[['Firstname','Lastname']] = patientheartrate['Name'].str.split(expand = True)
patientheartrate = patientheartrate.drop('Name',axis = 1)
print(patientheartrate)
```

```python
weight = patientheartrate['Weight']
for i in range(0,len(weight)):
  x = str(weight[i])
  if 'lbs' in x[-3:]:
    x = x[:-3:]
    float_x = float(x)
    y = int(float_x/2.2)
    y = str(y) + 'kgs'
    weight[i] = y
print(patientheartrate)
```

```python
patientheartrate.dropna(how="all", inplace = True)
print(patientheartrate)
```

```python
patientheartrate = patientheartrate.drop_duplicates(subset=["Firstname", "Lastname", "Age", "Weight"])
print(patientheartrate)
```

```python
patientheartrate.Firstname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
patientheartrate.Lastname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
print(patientheartrate)
```

```python
# vd7 lấy index của các dòng thiếu thông tin age và weight

index_ageweightnull = patientheartrate[patientheartrate['Age'].isna() & patientheartrate['Weight'].isna()].index
index_ageweightnull
```

```python
patientheartrate.drop(index_ageweightnull, inplace=True)
```

```python
patientheartrate["Age"].fillna(patientheartrate["Age"].mean(), inplace=False)
```

```python
# Vấn đề 8
#Melt the sex + time range columns in single column
patientheartrate = pd.melt(patientheartrate, id_vars=["ID","Age","Weight","Firstname","Lastname"],value_name="PulseRate", var_name="sex_and_time").sort_values(["ID","Age","Weight","Firstname","Lastname"])

# Extract sex, hour lower bound and hour upper bound group
tmp_patientheartrate = patientheartrate["sex_and_time"].str.extract("(\D)(\d+)(\d{2})", expand=True)
# Name columns
tmp_patientheartrate.columns = ["sex", "hours_lower", "hours_upper"]

# Create time columns based on "hours_lower" and "hours_upper" columns
tmp_patientheartrate["Time"] = tmp_patientheartrate["hours_lower"] + "." + tmp_patientheartrate["hours_upper"]

# Merge

patientheartrate = pd.concat([patientheartrate, tmp_patientheartrate], axis = 1)

# Drop unnecessary columns and rows
patientheartrate = patientheartrate.drop(["sex_and_time", "hours_lower", "hours_upper"], axis = 1)
patientheartrate = patientheartrate.dropna()
patientheartrate.to_csv("outputcleanup.csv", index=False)
print(patientheartrate)
```

```python
# Thay thế dấu gạch ngang bằng NaN để dễ dàng xử lý
patientheartrate['PulseRate'] = patientheartrate['PulseRate'].replace('-', np.nan)

# Tính số lượng và tỉ lệ phần trăm các giá trị bị thiếu
missing_count = patientheartrate['PulseRate'].isna().sum()
total_count = len(patientheartrate)
missing_percentage = (missing_count / total_count) * 100

print(f"Số lượng giá trị bị thiếu: {missing_count}")
print(f"Tỉ lệ phần trăm giá trị bị thiếu: {missing_percentage:.2f}%")

def fill_missing_pulse_rate(group):
    pulse_rate = group['PulseRate'].astype(float)

    # Phương pháp 1: Thay thế bằng giá trị trung bình liền trước và liền sau
    for i in range(len(pulse_rate)):
        if np.isnan(pulse_rate.iloc[i]):
            if i > 0 and i < len(pulse_rate) - 1 and not np.isnan(pulse_rate.iloc[i - 1]) and not np.isnan(pulse_rate.iloc[i + 1]):
                pulse_rate.iloc[i] = (pulse_rate.iloc[i - 1] + pulse_rate.iloc[i + 1]) / 2

    # Phương pháp 2: Thay thế bằng giá trị trung bình 2 giá liền trước
    for i in range(len(pulse_rate)):
        if np.isnan(pulse_rate.iloc[i]):
            if i > 1 and not np.isnan(pulse_rate.iloc[i - 1]) and not np.isnan(pulse_rate.iloc[i - 2]):
                pulse_rate.iloc[i] = (pulse_rate.iloc[i - 1] + pulse_rate.iloc[i - 2]) / 2

    # Phương pháp 3: Thay thế bằng giá trị trung bình 2 giá liền sau
    for i in range(len(pulse_rate)):
        if np.isnan(pulse_rate.iloc[i]):
            if i < len(pulse_rate) - 2 and not np.isnan(pulse_rate.iloc[i + 1]) and not np.isnan(pulse_rate.iloc[i + 2]):
                pulse_rate.iloc[i] = (pulse_rate.iloc[i + 1] + pulse_rate.iloc[i + 2]) / 2

    # Phương pháp 4: Trung bình của các giá trị huyết áp của người đó
    if pulse_rate.isna().sum() > 0:
        mean_individual = pulse_rate.mean()
        pulse_rate = pulse_rate.fillna(mean_individual)

    return pulse_rate

# Áp dụng hàm thay thế cho từng nhóm người
patientheartrate['PulseRate'] = patientheartrate.groupby('ID').apply(lambda group: fill_missing_pulse_rate(group)).reset_index(level=0, drop=True)

# Phương pháp 5: Trung bình của các giá trị huyết áp của nhóm giới tính
for sex in patientheartrate['sex'].unique():
    mean_sex = patientheartrate[patientheartrate['sex'] == sex]['PulseRate'].astype(float).mean()
    patientheartrate.loc[patientheartrate['sex'] == sex, 'PulseRate'] = patientheartrate.loc[patientheartrate['sex'] == sex, 'PulseRate'].fillna(mean_sex)

# Phương pháp 6: Trung bình của tất cả các giá trị huyết áp
mean_all = patientheartrate['PulseRate'].astype(float).mean()
patientheartrate['PulseRate'] = patientheartrate['PulseRate'].fillna(mean_all)

# Nếu vẫn còn giá trị thiếu, thay thế bằng mức ổn định trong y học (giả định là 70)
patientheartrate['PulseRate'] = patientheartrate['PulseRate'].fillna(70)

# In dữ liệu đã xử lý
print("Dữ liệu đã xử lý:")
print(patientheartrate.head())
```

```python
# Rút gọn dữ liệu, giữ lại các cột quan trọng
columns_to_keep = ['ID', 'Age', 'Weight', 'Firstname', 'Lastname', 'PulseRate', 'sex', 'Time']
patientheartrate_reduced = patientheartrate[columns_to_keep]

# Tái lập chỉ mục
patientheartrate_reduced = patientheartrate_reduced.reset_index(drop=True)

# Lưu dữ liệu đã xử lý thành tệp CSV mới với encoding 'utf-8'
patientheartrate_reduced.to_csv('patient_heart_rate_clean.csv', index=False, encoding='utf-8')

# In dữ liệu đã xử lý
print("Dữ liệu đã xử lý:")
print(patientheartrate_reduced.head())
```
