# **Phân tích cảm xúc người dùng**
Nhờ nhận xét và đánh giá của người dùng, các thuật toán phân tích cảm xúc có thể đánh giá mức
độ đánh giá hoặc đánh giá dựa trên các từ khóa nhất định. Mức độ đánh giá được gọi là ý kiến và
có ba giá trị có thể có: positive, neutral hoặc negative. Việc đánh giá ý kiến này do đó trở thành
một dạng phân loại. \
Mục đích của ví dụ dưới đây là tìm các từ lặp lại nhiều nhất dạng positive và dạng negative hoặc
các từ khóa liên quan đến một ý kiến nào đó, sử dụng thuật toán Naïve Bayes để phân loại cảm
xúc người dùng dựa vào các đoạn văn bản movie_reviews của họ. \
Trước tiên, download các movie review dùng lệnh sau:


```python
nltk.download('movie_reviews')
```

Sau đó xây dựng tập train dựa vào copus trên. Tạo 1 mảng tên là documents. Mảng này chứa cột
đầu tiên là nội dung review của người dùng, và cột thứ 2 là cột đánh giá positive, neutral, or
negative. Sau đó trộn các dòng này ngẫn nhiên

```python
import random
reviews = nltk.corpus.movie_reviews
documents = [(list(reviews.words(fileid)), category)
for category in reviews.categories()
for fileid in reviews.fileids(category)]
random.shuffle(documents)
```

Xem nội dung review đầu tiên (dòng 0, cột 0)

```python
first_review = ' '.join(documents[0][0])
print(first_review)
```

Xem kết quả review đầu tiên (dòng 0, cột 1)

```python
documents[0][1]
```

Ta cần tạo bảng phân phối tần số các từ trong copus, bảng này cần chuyển sang dạng list, ta dùng
hàm list()

```python
all_words = nltk.FreqDist(w.lower() for w in reviews.words())
word_features = list(all_words)
```

Sau đó, bước tiếp theo là xác định một hàm để tính toán các đặc trưng, tức là những từ đủ quan
trọng để thiết lập ý kiến của một review

```python
def document_features(document, word_features):
  document_words = set(document)
  features = {}
  for word in word_features:
    features['{}'.format(word)] = (word in document_words)
  return features
```

Khi bạn định nghĩa hàm document_features(), bạn tạo 1 tập các documents

```python
featuresets = [(document_features(d,word_features), c) for (d,c) in documents]
len(featuresets)
```

Tạo tập train và tập test: 1500 dòng đầu dùng cho tập train và 500 dòng còn lại dùng cho tập test
để đánh giá độ chính xác của mô hình.

```python
train_set, test_set = featuresets[1500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

Dùng thuật toán Naïve Bayes để phân loại, dùng thư viện NLTK. Sau đó tính toán độ chính xác
của thuật toán

```python
train_set, est_set = featuresets[1500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, test_set))
```

Chúng ta đã hoàn tất việc phân tích, dưới đây các từ với trọng số lớn nhất của các review được
đánh giá là positive và negative

```python
classifier.show_most_informative_features(10)
```

4. Viết chương trình Python với thư viện NLTK để loại bỏ các stopword từ một văn bản đã cho.

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stopwords(text):
  stop_words = set(stopwords.words('english'))
  words = nltk.word_tokenize(text)
  filtered_words = [word for word in words if word.lower() not in stop_words]
  return ' '.join(filtered_words)

text = input('Nhập vào văn bản: ')
filtered_text = remove_stopwords(text)
print("\nVăn bản gốc:", text)
print("\nVăn bản sau khi loại bỏ stopword:", filtered_text)
```

5. Viết chương trình Python với thư viện NLTK bỏ qua các stopword từ danh sách các stopword.

```python
def ignore_stopwords(text, ignored_words):
  stop_words = set(stopwords.words('english')) - set(ignored_words)
  words = nltk.word_tokenize(text)
  filtered_words = [word for word in words if word.lower() not in stop_words]
  return ' '.join(filtered_words)

text = input('Nhập vào văn bản: ')
ignored_words = input('\nNhập vào các stopword bạn muốn bỏ qua (phân cách bằng dấu phẩy): ').split(',')
filtered_text = ignore_stopwords(text, ignored_words)
print("\nVăn bản gốc:", text)
print("\nVăn bản sau khi bỏ qua các stopword:", filtered_text)
```

6. Viết một chương trình Python với thư viện NLTK để tìm định nghĩa và ví dụ của một từ đã cho bằng WordNet từ Wikipedia.

```python
!pip install wikipedia-api
```

```python
import nltk
from nltk.corpus import wordnet as wn
import wikipediaapi

# Tải xuống dữ liệu NLTK cần thiết
nltk.download('wordnet')
nltk.download('omw-1.4')

def get_wordnet_info(word):
    synsets = wn.synsets(word)
    if not synsets:
        return None

    definitions = []
    examples = []

    for synset in synsets:
        definitions.append(synset.definition())
        examples.extend(synset.examples())

    return definitions, examples

def get_wikipedia_info(word):
    user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    wiki_wiki = wikipediaapi.Wikipedia('en', headers={'User-Agent': user_agent})
    page = wiki_wiki.page(word)

    if not page.exists():
        return None

    return page.summary

def main(word):
    # Lấy định nghĩa và ví dụ về WordNet
    wordnet_info = get_wordnet_info(word)
    if wordnet_info:
        definitions, examples = wordnet_info
        print(f"Định nghĩa WordNet cho '{word}':")
        for i, definition in enumerate(definitions, 1):
            print(f"{i}. {definition}")

        print(f"\nVí dụ về WordNet cho '{word}':")
        for i, example in enumerate(examples, 1):
            print(f"{i}. {example}")
    else:
        print(f"Không tìm thấy thông tin WordNet cho '{word}'.")

    # Lấy bản tóm tắt Wikipedia
    wikipedia_info = get_wikipedia_info(word)
    if wikipedia_info:
        print(f"\nTóm tắt Wikipedia cho '{word}':")
        print(wikipedia_info)
    else:
        print(f"Không tìm thấy thông tin Wikipedia cho '{word}'.")

if __name__ == "__main__":
    word = input("\nNhập vào từ khóa: ")
    main(word)
```

7. Viết chương trình Python với thư viện NLTK để tìm tập hợp các từ đồng nghĩa và trái nghĩa
của một từ nào đó.

```python
nltk.download('wordnet')

def find_synonyms_antonyms(word):
    synonyms = []
    antonyms = []
    for synset in wn.synsets(word):
        for lemma in synset.lemmas():
            synonyms.append(lemma.name())
            if lemma.antonyms():
                antonyms.append(lemma.antonyms()[0].name())
    return synonyms, antonyms

if __name__ == "__main__":
    word = input("Nhập vào từ khóa: ")
    synonyms, antonyms = find_synonyms_antonyms(word)
    print(f"Từ đồng nghĩa của '{word}': {set(synonyms)}")
    print(f"Từ trái nghĩa của '{word}': {set(antonyms)}")
```

9. Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai danh từ đã
cho.

```python
nltk.download('wordnet')

def compare_noun_similarity(noun1, noun2):
  synset1 = wn.synsets(noun1, pos=wn.NOUN)
  synset2 = wn.synsets(noun2, pos=wn.NOUN)

  if not synset1 or not synset2:
    return 0

  max_similarity = 0
  for s1 in synset1:
    for s2 in synset2:
      similarity = s1.path_similarity(s2)
      if similarity is not None and similarity > max_similarity:
        max_similarity = similarity

  return max_similarity

if __name__ == "__main__":
  noun1 = input("Nhập danh từ thứ nhất: ")
  noun2 = input("Nhập danh từ thứ hai: ")

  similarity = compare_noun_similarity(noun1, noun2)
  print(f"Độ giống nhau giữa '{noun1}' và '{noun2}' là: {similarity}")
```

10. Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai động từ đã
cho.


```python
def compare_verb_similarity(verb1, verb2):
  synset1 = wn.synsets(verb1, pos=wn.VERB)
  synset2 = wn.synsets(verb2, pos=wn.VERB)

  if not synset1 or not synset2:
    return 0

  max_similarity = 0
  for s1 in synset1:
    for s2 in synset2:
      similarity = s1.path_similarity(s2)
      if similarity is not None and similarity > max_similarity:
        max_similarity = similarity

  return max_similarity

if __name__ == "__main__":
  verb1 = input("Nhập động từ thứ nhất: ")
  verb2 = input("Nhập động từ thứ hai: ")

  similarity = compare_verb_similarity(verb1, verb2)
  print(f"Độ giống nhau giữa '{verb1}' và '{verb2}' là: {similarity}")
```

```python
# Đọc tên nam và nữ từ kho ngữ liệu
male_names = nltk.corpus.names.words('male.txt')
female_names = nltk.corpus.names.words('female.txt')

# Tạo danh sách các cặp tên (nam, nữ)
name_pairs = [(male, female) for male in male_names for female in female_names]

# Trộn ngẫu nhiên danh sách các cặp tên
random.shuffle(name_pairs)

# In 15 cặp tên ngẫu nhiên đầu tiên
print("15 cặp tên ngẫu nhiên (nam, nữ):")
for i in range(15):
    male, female = name_pairs[i]
    print(f"- {male}, {female}")
```

13. Viết chương trình Python với thư viện NLTK để trích xuất ký tự cuối cùng của tất cả các tên
được gắn nhãn và tạo mảng mới với chữ cái cuối cùng của mỗi tên và nhãn được liên kết.

```python
# Đọc tên nam và nữ từ kho ngữ liệu
male_names = nltk.corpus.names.words('male.txt')
female_names = nltk.corpus.names.words('female.txt')

# Tạo mảng mới với chữ cái cuối cùng của mỗi tên và nhãn được liên kết
labeled_last_letters = []
for name in male_names:
    labeled_last_letters.append((name[-1], 'male'))
for name in female_names:
    labeled_last_letters.append((name[-1], 'female'))

# In mảng mới
print("Mảng mới với chữ cái cuối cùng của mỗi tên và nhãn được liên kết:")
for last_letter, label in labeled_last_letters:
    print(f"- {last_letter}: {label}")
```

```python
# Hàm để trích xuất ký tự cuối cùng và tạo mảng mới
def extract_last_character_with_label(name_list, label):
    last_characters = []
    for name in name_list:
        if name:
            last_char = name[-1]  # Lấy ký tự cuối cùng của tên
            last_characters.append((last_char, label))  # Thêm vào mảng với nhãn tương ứng
    return last_characters

# Trích xuất ký tự cuối cùng và nhãn từ danh sách các tên
male_last_chars = extract_last_character_with_label(male_names, 'male')
female_last_chars = extract_last_character_with_label(female_names, 'female')

# Kết hợp các mảng
all_last_chars = male_last_chars + female_last_chars

# In kết quả
print("Ký tự cuối cùng và nhãn tương ứng:")
for char, label in all_last_chars:
    print(f"{char} - {label}")
```
